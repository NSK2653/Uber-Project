```python
import pandas as pd
df=pd.read_excel(r"C:\Users\mohan-14784\Downloads\uber data (3).xlsx")
df.head()
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'], dayfirst=True, errors='coerce')
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'], dayfirst=True, errors='coerce')
df['Hour'] = df['Request timestamp'].dt.hour
df['Date'] = df['Request timestamp'].dt.date
df['Hour'].value_counts().sort_index().plot(kind='bar', title='Requests per Hour')
df['Status'].value_counts().plot(kind='bar', title='Status Distribution')
gap = df[df['Status'].isin(['Cancelled', 'No Cars Available'])]
gap['Pickup point'].value_counts().plot(kind='bar', title='Demand Gaps by Pickup Point')
df['Date'].value_counts().sort_index().plot(kind='line', title='Requests per Day')
```
